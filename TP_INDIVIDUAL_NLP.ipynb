### Info Básica

Configuración de  Selenium para usar Chrome en modo headless, Chrome se ejecutará en "modo headless" (sin interfaz gráfica).

```python
# Configuración del navegador
chrome_options = Options()
chrome_options.add_argument("--headless")
chrome_options.add_argument("--no-sandbox")
chrome_options.add_argument("--disable-dev-shm-usage")

# Inicializar el driver
driver = webdriver.Chrome(options=chrome_options)

# Abrir la página
driver.get('https://boardgamegeek.com/boardgame/312484/lost-ruins-of-arnak')
```

```python
# Espera explícita para asegurar que el título esté presente
wait = WebDriverWait(driver, 20)  # Esperar hasta 10 segundos
title_element = wait.until(EC.presence_of_element_located((By.CSS_SELECTOR, 'h1 a span[itemprop="name"]')))
title_game = title_element.text
print(f"Título del juego: {title_game}")

# Obtener el número de jugadores
players_element = wait.until(EC.presence_of_element_located((By.CSS_SELECTOR, 'ul.gameplay li[itemscope][itemprop="numberOfPlayers"] p.gameplay-item-primary')))
players_game = players_element.text
print(f"Número de jugadores: {players_game}")

# Obtener la edad mínima recomendada
min_age_element = wait.until(EC.presence_of_element_located((By.CSS_SELECTOR, 'span[itemprop="suggestedMinAge"]')))
min_age_game = min_age_element.text
print(f"Edad mínima recomendada: {min_age_game}")

# Obtener los diseñadores
designer_element = wait.until(EC.presence_of_element_located((By.CSS_SELECTOR, 'li.ng-scope popup-list[items*="geekitemctrl.geekitem.data.item.links"]')))
designer_game = designer_element.text
print(f"Diseñadores: {designer_game}")

# Obtener el artista
artist_element = wait.until(EC.presence_of_element_located((By.CSS_SELECTOR, "a[href*='/boardgameartist/'] span.ng-binding")))
artist_game = artist_element.text
print(f"Artista: {artist_game}")

# Obtener la duración del juego
time_container = wait.until(EC.presence_of_element_located((By.XPATH, "//li[h3[contains(text(), 'Play Time')]]")))
duration_text = time_container.text.strip()
lines = duration_text.split('\n')
duration_game = lines[1]
print(f"Duración del juego: {duration_game}")
```

### Info foro

```python
driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()), options=chrome_options)
# Crear un nuevo documento Word
doc = Document()

# Función para scrapear los hilos individuales
def get_thread_details(thread_url):
    driver.get(thread_url)
    time.sleep(3)  # Esperar a que se cargue el contenido dinámico
    html = driver.page_source
    soup = BeautifulSoup(html, 'html.parser')

    # Buscar los comentarios dentro de las etiquetas <gg-markup-safe-html>
    comments = soup.find_all('gg-markup-safe-html')
    thread_content = ""

    for comment in comments:
        thread_content += comment.get_text(separator="\n", strip=True) + "\n\n"

    return thread_content

# Loop para iterar sobre varias páginas
for id in [1, 2, 3, 4]:
    url = f'https://boardgamegeek.com/boardgame/312484/lost-ruins-of-arnak/forums/0?pageid={id}'
    driver.get(url)
    time.sleep(3)

    # Obtener el HTML completo de la página cargada
    html = driver.page_source
    soup = BeautifulSoup(html, 'html.parser')

    # Seleccionar todos los <li> con la clase 'summary-item ng-scope'
    li_items = soup.find_all('li', class_='summary-item ng-scope')

    # Iterar sobre cada elemento <li>
    for li in li_items:
        # Extraer el título
        title = li.find('h3', class_='m-0 fs-sm text-inherit leading-inherit text-inline')
        if title:
            title_text = title.get_text(strip=True)
            doc.add_paragraph(f"Título: {title_text}")

        # Extraer el enlace del hilo
        link = li.find('a', {'ng-href': True})
        if link:
            thread_url = "https://boardgamegeek.com" + link['ng-href']

            # Obtener los detalles del hilo (comentarios)
            thread_details = get_thread_details(thread_url)
            doc.add_paragraph(thread_details)

        # Agregar un salto de página después de cada hilo
        doc.add_paragraph('')

# Guardar el documento Word con el contenido scrapeado
doc.save('comentarios.docx')
print("Documento guardado como 'comentarios.docx'")

# Cerrar el navegador al final
driver.quit()
```

```python
reglas_turno_juego = "lost-ruins-of-arnak-player-aid-es.pdf"

# Leer el archivo PDF
reader = PdfReader(reglas_turno_juego)

# Extraer texto del PDF
turno_juego_texto = ""
for page in reader.pages:
    turno_juego_texto += page.extract_text()

# Reemplazar caracteres no deseados
turno_juego_texto = turno_juego_texto.replace('\uf0a7', '')

# Mostrar el texto extraído
print(turno_juego_texto)
```

### Limpieza del texto

```python
import re
# Paso 1: Limpiar caracteres extraños y saltos de línea innecesarios
texto_limpio = turno_juego_texto.replace('\uf0a7', '').replace('\n ', '\n')

# Paso 2: Reemplazar bullets mal formateados con guiones para listas
texto_limpio = re.sub(r'•\s*', '- ', texto_limpio)

# Paso 3: Eliminar espacios aleatorios entre letras y palabras
texto_limpio = re.sub(r'(\b[A-Z])\s+', r'\1', texto_limpio)

# Paso 4: Mejorar estructura por secciones con títulos
# Agregar un salto de línea antes de títulos principales (opcional)
texto_limpio = re.sub(r'(Turno de juego|Preparación para la siguiente ronda|Jerarquía de desplazamientos|Efectos)', r'\n\1', texto_limpio)

# Paso 5: Eliminar espacios repetidos
texto_limpio = re.sub(r'\n\s*\n', '\n', texto_limpio)

# Resultado formateado
print(texto_limpio)
```

##  Creación de Base de Datos Vectorial

PARA INFORME:


Texto a Vectores: Utilizas un modelo de embedding (como SentenceTransformer o KeyBERT) para convertir los chunks de texto en vectores numéricos.


Almacenamiento de Vectores: Estos vectores, junto con metadatos relevantes (por ejemplo, palabras clave, idioma, etc.), se almacenan en ChromaDB, que es una base de datos diseñada para manejar y gestionar estos vectores.


Búsqueda y Comparación: Luego, puedes realizar búsquedas basadas en la similitud de los vectores. Por ejemplo, si deseas encontrar los textos más similares a una consulta dada, puedes calcular el vector de la consulta y buscar los vectores más cercanos en la base de datos.

```python
# Extracción Automática de Palabras Clave y Almacenamiento de Chunks en ChromaDB

from langchain.text_splitter import RecursiveCharacterTextSplitter
from sentence_transformers import SentenceTransformer
from chromadb import Client
from chromadb.config import Settings
from keybert import KeyBERT
import uuid
import os

# Ruta a los textos limpios generados previamente
text_files = ["turno_juego_limpio.txt", "resena.txt", "resena2.txt"]

# Función para leer los archivos de texto
def read_text_files(file_paths):
    texts = []
    for file_path in file_paths:
        with open(file_path, 'r', encoding='utf-8') as f:
            texts.append(f.read())
    return texts

# Función para dividir el texto en chunks
def split_texts(texts, chunk_size=500, chunk_overlap=50):
    splitter = RecursiveCharacterTextSplitter(
        chunk_size=chunk_size, chunk_overlap=chunk_overlap
    )
    chunks = []
    for text in texts:
        chunks.extend(splitter.split_text(text))
    return chunks

# Función para limpiar los chunks (eliminar espacios innecesarios y filtrar los muy pequeños)
def clean_chunks(chunks):
    cleaned_chunks = []
    for chunk in chunks:
        cleaned_chunk = chunk.strip()  # Eliminar espacios innecesarios
        if len(cleaned_chunk) > 10:  # Filtrar chunks muy pequeños
            cleaned_chunks.append(cleaned_chunk)
    return cleaned_chunks

# Función para obtener las palabras clave dinámicas con KeyBERT
def extract_keywords(text, kw_model, top_n=5):
    keywords_dynamic = [kw[0] for kw in kw_model.extract_keywords(text, top_n=top_n)]
    return keywords_dynamic

# Función para obtener palabras clave manuales basadas en el nombre del archivo
def get_keywords(filename):
    if "turno_juego" in filename.lower():
        return "turno, juego, reglas"
    elif "resena" in filename.lower():
        return "reseña, juego, opiniones"
    else:
        return "general"

# Función para determinar el idioma basado en el nombre del archivo
def get_language_from_filename(filename):
    if "espanol" in filename.lower():
        return "es"
    elif "english" in filename.lower():
        return "en"
    else:
        return "multilingual"

# Crear el modelo de embeddings
embedding_model = SentenceTransformer('all-MiniLM-L6-v2')

# Crear el modelo KeyBERT para extraer palabras clave
kw_model = KeyBERT('all-MiniLM-L6-v2')

# Configuración de ChromaDB
client = Client(Settings(
    persist_directory="./chromadb",  # Directorio de persistencia
    anonymized_telemetry=False       # Desactivar telemetría si no es necesaria
))

# Crear o acceder a la colección de ChromaDB
collection = client.get_or_create_collection(name="text_chunks")

# Función para almacenar embeddings en ChromaDB
def store_in_chromadb(chunks, embeddings, collection_name="text_chunks"):
    # Obtener IDs existentes para evitar duplicados
    existing_ids = set(collection.get()["ids"])

    # Añadir embeddings a la colección con metadatos
    for i, (chunk, embedding) in enumerate(zip(chunks, embeddings)):
        doc_id = f"chunk_{i}"
        if doc_id not in existing_ids:  # Evitar duplicados
            metadata = {
                "chunk": chunk,
                "language": get_language_from_filename(file_path),
                "keywords": ", ".join(extract_keywords(chunk, kw_model))  # Palabras clave dinámicas
            }
            collection.add(
                embeddings=[embedding.tolist()],
                metadatas=[metadata],
                ids=[doc_id]
            )
    print(f"Stored {len(chunks)} chunks in ChromaDB collection '{collection_name}'.")

    return collection

# Visualizar datos almacenados en ChromaDB
def display_stored_data(collection):
    docs = collection.get()
    for i, doc in enumerate(docs['metadatas']):
        print(f"Chunk {i}: {doc['chunk'][:100]}...")  # Mostrar solo los primeros 100 caracteres

# Ejecución principal
if __name__ == "__main__":
    # Leer textos
    texts = read_text_files(text_files)

    # Dividir los textos en chunks
    chunks = split_texts(texts)

    # Limpiar los chunks
    cleaned_chunks = clean_chunks(chunks)

    # Crear embeddings
    embeddings = embedding_model.encode(cleaned_chunks, show_progress_bar=True)

    # Almacenar los embeddings en ChromaDB con metadatos
    store_in_chromadb(cleaned_chunks, embeddings)

    # Mostrar los datos almacenados
    display_stored_data(collection)

    print("Proceso completado.")
```

### Consultas




```python
def query_chromadb(collection, query_text, n_results=5):
    results = collection.query(
        query_texts=[query_text],
        n_results=n_results  # Número de resultados relevantes
    )
    return results

# Ejemplo de consulta
if __name__ == "__main__":
    query = "¿cuantas cartas son?"
    results = query_chromadb(collection, query)
    for i, (doc, score) in enumerate(zip(results['metadatas'], results['distances'])):
        score_value = score[0] if isinstance(score, list) else score

        # Imprime el contenido completo de doc
        print(f"Resultado {i+1} (Score: {score_value:.2f}):")
        print(doc)  # Para ver la estructura completa
        print("\n" + "-"*50)
```
